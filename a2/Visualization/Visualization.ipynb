```python
sub_category=list(set(df["Sub-Category"].tolist()))
sub_category=[str(i) for i in sub_category]
```

```python
sub_category
```

```python
profit=[]
for i in range(len(sub_category)):
    a= df[df["Sub-Category"]==sub_category[i]]["Profit"].sum()
    profit.append(a)
```

```python
profit
```

```python
fig = mpl.figure()
axes= fig.add_axes([0,0,2.3,1])
axes.bar(sub_category,profit,color="R")
mpl.xlabel("(Sub category)")
mpl.ylabel("Profit")
```

```python
unique_category =list(set(df["Category"].tolist()))
profit1=[]
for i in range(len(unique_category)):
    var= df[df["Category"] == unique_category[i]]["Profit"].sum()
    profit1.append(var)
```

```python
date_list=df["Order Date"].tolist()
profit=df["Profit"].tolist()
date_list=[str(i) for i in date_list]
```

```python
import re
year_list=[]
for i in range(len(date_list)):
    var=int(re.findall("[0-9]{4}",date_list[i])[0])
    year_list.append(var)
```

```python
year_list1 = list(set(year_list))
yearwise_profit=[]
for i in range(len(year_list1)):
    sum1=0
    for j in range(len(year_list)):
        if(year_list1[i]==year_list[j]):
            sum1=sum1+profit[j]
    yearwise_profit.append(sum1)
```

```python
yearwise_profit
```

```python
year_list1
```

```python
mpl.plot(year_list1,yearwise_profit)
mpl.xlabel("Year")
mpl.ylabel("Profit")
```

```python
############## 5.f ) 
```

```python
cust_id=list(set(df["Customer ID"].tolist()))

final=[]
for i in range(len(cust_id)):
    var= df[df["Customer ID"] == cust_id[i]]["Profit"].sum()
    temp=[cust_id[i],var]
    final.append(temp)

final.sort(key = lambda x: x[1] , reverse= True)
```

```python
print("Top 10 most profitable customers are:")
for i in final[:10]:
    print(i[0],"=> ",i[1])
```

```python
Dept = pd.DataFrame([[10,"Police","Bangalore"],[11,"Police1","Kolkata"],[12,"Police2","Hyderabad"],[13,"Police3","Delhi"],[14,"Police4","Patna"]], columns=["Deptno","Dname","Loc"])
```

```python
Emp = pd.DataFrame([[10,100,200000],[21,101,120000],[12,102,130000],[23,104,140000],[14,105,300000]], columns=["Deptno","Employee Number","Salary"])
```

```python
Dept
```

```python
Emp
```

```python
# inner join
merged_inner = pd.merge(left=Dept,right=Emp, left_on='Deptno', right_on='Deptno')
```

```python
merged_inner
```

```python
# leftouter join
merged_left = pd.merge(left=Dept,right=Emp,how='left', left_on='Deptno', right_on='Deptno')
```

```python
merged_left
```

```python
# rightouter join
merged_right = pd.merge(left=Dept,right=Emp,how='right', left_on='Deptno', right_on='Deptno')
```

```python
merged_right
```

```python
# Fullouter join
merged_full = pd.merge(left=Dept,right=Emp,how='outer', left_on='Deptno', right_on='Deptno')
```
