```python
import pandas as pd
```

```python
import numpy as np
```

```python
depts=np.random.randint(2210,2215,size=15)
```

```python
emps=list(range(2222,2237))
```

```python
sals=np.random.randint(100000,500000,size=15)
```

```python
emp_data_dict={'DeptNo':depts,
                'EmpNo':emps,
                'Sal':sals}
```

```python
emp_data=pd.DataFrame(emp_data_dict)
```

```python
emp_data
```

```python
def mapper(dept):
    if dept==2210:
        return 'IT'
    if dept==2211:
        return 'Finanace'
    if dept==2212:
        return 'Sec'
    if dept==2213:
        return 'FMG'
    if dept==2214:
        return 'IMG'
```

```python
depts=[2211,2212,2213,2214,2215,2216,2217]
```

```python
dept_data_dict={'DeptNo':depts,'DeptNm':list(map(mapper,depts)),
              'location':np.concatenate([2*['Hyd'],2*['Banglr'],3*['Mumbai']]).ravel()}
```

```python
dept_data_dict
```

```python
depts_data=pd.DataFrame(dept_data_dict)
```

```python
depts_data
```

```python
#Inner join
emp_data.merge(depts_data)
```

```python
#Left outer join
emp_data.merge(depts_data,how='left')
```

```python
#Right outer join
emp_data.merge(depts_data,how='right')
```

```python
#Full outer join
emp_data.merge(depts_data,how='outer')
```
